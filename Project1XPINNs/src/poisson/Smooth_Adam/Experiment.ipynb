```python
import numpy as onp 
import matplotlib.pyplot as plt

path = "pinn"
```

```python
losses = []
rel_l2 = []

for i in range(1,11):
    file = path + f"/poisson_{i}.npz"
    b = onp.load(file)
    losses.append(b["arr_0"])
    print(len(b["arr_0"]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

iterations = np.linspace(1000, 200000, 200)
print(iterations)
# Compute statistics
medians = np.median(losses, axis=0)
q1 = np.percentile(losses, 25, axis=0)
q3 = np.percentile(losses, 75, axis=0)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(iterations,medians, color='blue', label='Median')
plt.fill_between(iterations, q1, q3, color='blue', alpha=0.2, label='Interquartile Range')
plt.xlabel('Entry')
plt.ylabel('Value')
plt.yscale('log')
plt.title('Median with Interquartile Range')
plt.legend()
plt.grid(True)
plt.show()
```
