# Poisson Points single domain (only one PINN)

```python
from region_utils.region import Subdomain, Domain
from region_utils.shapes import ConvexPolygon
import jax.numpy as np
import matplotlib.pyplot as plt
```

```python
vertecies = np.array([[0, 0], [0,1], [1, 1], [1, 0]]).reshape(4,-1)
square = ConvexPolygon(
   vertecies,
    [0, 1, 2, 3],
)

subdom = Subdomain([square])
domain = Domain([subdom])
```

```python
### Create domain after paper specifications https://epubs.siam.org/doi/epdf/10.1137/21M1447039
domain.create_interior(900, [0, 0], [1, 1])
domain.create_boundary(120)
domain.plot()
```

```python
from utils import data_path

domain.write_to_file(data_path / "poisson_single_pinn_smooth_train.json") #####OBSSSSSSSS
```

```python
domain.create_testing_data(100, [0, 0], [1, 1])
domain.write_to_file(data_path / "poisson_single_pinn_smooth_test.json", False) #######OBBBBBSSSSS
domain.plot(False)
```
