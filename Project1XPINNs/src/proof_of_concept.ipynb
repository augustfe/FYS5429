```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
regions = [[] for i in range(3)]
```

```python
x_vals = np.random.uniform(-1, 1, (2000, 1))
t_vals = np.random.uniform(0, 1, (2000, 1))
```

```python
X = np.hstack([x_vals, t_vals])
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
regions = [[] for i in range(3)]

for x, t in X:
    if -0.2 >= x - 0.5 * t:
        regions[0].append([x, t])
    elif x - 0.5 * t >= 0.2:
        regions[1].append([x, t])
    else:
        regions[2].append([x, t])

for i, sub_list in enumerate(regions):
    regions[i] = np.array(regions[i])

for i, sub_list in enumerate(regions):
    plt.scatter(sub_list[:, 0], sub_list[:, 1], s=2.5, label=f"Region {i+1}")
plt.legend()
```
